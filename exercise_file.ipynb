```python
# initialize circuit with 1 qbit and 1 classical bit
circuit = QuantumCircuit(1,1)
circuit.x(0)
circuit.h(0)
circuit.h(0)
#circuit.ry(pi/4, [0,1])
#circuit.z(1)
#print(circuit)
circuit.draw(output = 'mpl');
```
